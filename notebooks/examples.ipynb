```python
import sys
sys.path.append('..')

import os

import cv2
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets.widgets import interact
import ipywidgets as widgets

from visualiser import SliceMask
```

```python
def plot_image(img):
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    ax.set_xticks([])
    ax.set_yticks([])

    plt.imshow(img, cmap=plt.cm.bone)
    plt.plot()
```

```python
mask_obj = SliceMask.from_json('../data/I1450.png.json')
print(mask_obj.size)
print(len(mask_obj.contours))
```

```python
plot_image(mask_obj.mask)
```

```python
plot_image(mask_obj.contour_mask)
```

```python
numpy_mask = mask_obj.mask
numpy_mask = cv2.circle(numpy_mask, (120, 100), 30, (1,), -1)

mask_obj2 = SliceMask.from_array(numpy_mask.copy())
print(len(mask_obj2.contours))
plot_image(mask_obj2.mask)
```
